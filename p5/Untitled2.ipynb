```python
import decimal
import math
import matplotlib.pyplot as plt
decimal.getcontext().prec = 102
```

```python

def answer(n):
    val = decimal.Decimal(n)
    sq2 = decimal.Decimal(2).sqrt()
    
    q = 1
    x = []
    y = []
    xs,ys = [],[]
    p = 0
    cnt, count = 0,1
    while q <= val:
        v = (q * math.sqrt(2))//1.0
        xs += [q]
        ys += [v]
        if p == 0:
            p = v
            q += 1
            continue
#         print(v,p, q)
        if v - p == 1:
            count += 1
        else:
#             print(cnt, count)
            x += [cnt]
            cnt += 1
            y += [count]
            count = 1
            
        
        p = v
        q += 1
#     plt.rcParams["figure.figsize"] = (20,20)
#     plt.scatter(xs,ys)
#     plt.show()
    plt.rcParams["figure.figsize"] = (20,3)
    plt.plot(x,y)
    plt.show()
    print(y)




answer("200")

```
